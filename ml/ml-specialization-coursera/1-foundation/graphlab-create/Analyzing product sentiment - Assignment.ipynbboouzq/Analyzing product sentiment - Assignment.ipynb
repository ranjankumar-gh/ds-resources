```python
import graphlab
```

```python
products = graphlab.SFrame('amazon_baby_assignment.gl/')
```

```python
products.head()
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate'];
```

```python
def f1(review_text):
    selected_words_dict = {}
    for f in selected_words:
        selected_words_dict.update([(f,0)])
        
    review_splitted = review_text.lower().split()
    for words in review_splitted:
        if selected_words_dict.has_key(words):
            selected_words_dict[words] = selected_words_dict[words] + 1
    return selected_words_dict

    
products['selected_words'] = products['review'].apply(lambda review_text: f1(review_text))
#products['awesome'] = products['review'].apply(lambda review_text: review_text.count('awesome'))
products['awesome'] = products['selected_word_count'].apply(lambda x: x['awesome'])
products['great'] = products['selected_word_count'].apply(lambda x: x['great'])
products['fantastic'] = products['selected_word_count'].apply(lambda x: x['fantastic'])
products['amazing'] = products['selected_word_count'].apply(lambda x: x['amazing'])
products['love'] = products['selected_word_count'].apply(lambda x: x['love'])
products['horrible'] = products['selected_word_count'].apply(lambda x: x['horrible'])
products['bad'] = products['selected_word_count'].apply(lambda x: x['bad'])
products['terrible'] = products['selected_word_count'].apply(lambda x: x['terrible'])
products['awful'] = products['selected_word_count'].apply(lambda x: x['awful'])
products['wow'] = products['selected_word_count'].apply(lambda x: x['wow'])
products['hate'] = products['selected_word_count'].apply(lambda x: x['hate'])

#products['selected_word_count'].show()
graphlab.canvas.set_target('ipynb')

print "awesome: %d" % sum(products['awesome'])
print "great: %d" % sum(products['great'])
print "fantastic: %d" % sum(products['fantastic'])
print "amazing: %d" % sum(products['amazing'])
print "love: %d" % sum(products['love'])
print "horrible: %d" % sum(products['horrible'])
print "bad: %d" % sum(products['bad'])
print "terrible: %d" % sum(products['terrible'])
print "awful: %d" % sum(products['awful'])
print "wow: %d" % sum(products['wow'])
print "hate: %d" % sum(products['hate'])

products
```

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
selected_words_model = graphlab.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=['selected_words'],
                                                     validation_set=test_data)
```
