## Selecting
You can think of a DataFrame as a group of Series that share an index (in this case the column headers). This makes it easy to select specific columns.

Selecting a single column from the DataFrame will return a Series object.

```python
sales['sqft_living'].head()
```

To select multiple columns, simply pass a list of column names to the DataFrame, the output of which will be a DataFrame.

```python
print sales[['sqft_living', 'price']].head()
print '\n'

# can also store in a variable to use later
columns_you_want = ['bedrooms', 'price'] 
print sales[columns_you_want].head()
```

Row selection can be done multiple ways, but doing so by an individual index or boolean indexing are typically easiest.

```python
print sales[sales.bedrooms > 2].head(3)
print '\n'

print sales[(sales.bathrooms == 1) & (sales.bedrooms == 2)].head(3)
print '\n'

print sales[(sales.waterfront == 0) | (sales.bedrooms < 3)].head(3)
```

```python
print sales.set_index('id').head()
print '\n'

print sales.head()
print "\n^^^ I didn't actually change the DataFrame. ^^^\n"

with_new_index = sales.set_index('id')
print with_new_index.head()
print "\n^^^ set_index actually returns a new DataFrame. ^^^\n"

```

If you want to modify your existing DataFrame, use the inplace parameter.

```python
sales.set_index('id', inplace=True)
sales.head()
```

We've lost the default pandas 0-based index and moved the id into its place. We can select rows based on the index using the ix method.

```python
print sales.ix[7129300520]
print '\n'
print sales.ix[[2487200875, 5631500400, 6414100192]]
```

If we want old pandas default index back, we can just reset_index. The same rules for inplace apply.

```python
sales.reset_index(inplace=True)
sales.head()
```

**Joining**, **Combining**, **Grouping**, can be used with DataFrames.
