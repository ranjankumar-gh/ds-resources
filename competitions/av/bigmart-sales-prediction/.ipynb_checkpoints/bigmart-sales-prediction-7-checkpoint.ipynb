## Scaling the data

```python
sc = StandardScaler()
X = sc.fit_transform(X)
X_test = sc.transform(X_test)
```

## Dividing the data in training and validation set

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Training

```python
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
```

## Predictions

```python
predictions = model_lr.predict(X_val)
print(model_lr.score(X_val, y_val))
model_lr.score(X_train, y_train)
```

```python
model_lr = LinearRegression()
model_lr.fit(X, y)
```

```python
def lin_regplot(X, y, model):
    plt.scatter(X, y, c='blue')
    plt.plot(X, model.predict(X), color='red')
    return None
```

```python
lin_regplot(X, y, model_lr)
```

```python
print('Slope: %.3f' % model_lr.coef_[0])
print('Slope: %.3f' % model_lr.coef_[1])
print('Intercept: %.3f' % model_lr.intercept_)
```
