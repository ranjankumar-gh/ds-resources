# Applying K-Nearest Mode for Loan prediction

```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
loans = pd.read_csv('train_u6lujuX.csv')
test_data = pd.read_csv('test_Y3wMUE5.csv')
loans.head()
```

## Preparing the data

### Check if values are missing in the data point

Self_Employed

### Analysing Income

```python
print len(loans)

x = np.arange(0, len(loans['ApplicantIncome'])) 
y = loans['ApplicantIncome']

ll = plt.plot(x,y)
plt.show()
```

## Transform data categories to numerical categories
Training Data

```python
loans['Loan_Status'] = loans['Loan_Status'].apply(lambda x: 1 if x=='Y' else 0)
loans['Self_Employed'] = loans['Self_Employed'].apply(lambda x: 1 if x=='Yes' else 0)
loans['ApplicantIncome_gt_30k'] = loans['ApplicantIncome'].apply(lambda x: 1 if x >= 30000 else 0)
loans['ApplicantIncome_bwn_20k_30k'] = loans['ApplicantIncome'].apply(lambda x: 1 if x < 30000 and x >= 20000 else 0)
loans['ApplicantIncome_bwn_10k_20k'] = loans['ApplicantIncome'].apply(lambda x: 1 if x < 20000 and x >= 10000 else 0)
loans['ApplicantIncome_lt_10k'] = loans['ApplicantIncome'].apply(lambda x: 1 if x < 10000 else 0)
loans.head()
```

Test Data

```python
test_data['Self_Employed'] = test_data['Self_Employed'].apply(lambda x: 1 if x=='Yes' else 0)
test_data['ApplicantIncome_gt_30k'] = test_data['ApplicantIncome'].apply(lambda x: 1 if x >= 30000 else 0)
test_data['ApplicantIncome_bwn_20k_30k'] = test_data['ApplicantIncome'].apply(lambda x: 1 if x < 30000 and x >= 20000 else 0)
test_data['ApplicantIncome_bwn_10k_20k'] = test_data['ApplicantIncome'].apply(lambda x: 1 if x < 20000 and x >= 10000 else 0)
test_data['ApplicantIncome_lt_10k'] = test_data['ApplicantIncome'].apply(lambda x: 1 if x < 10000 else 0)
test_data.head()
```

## Training the model

```python
feature_cols = ['Self_Employed', 'ApplicantIncome']
X = loans[feature_cols]
y = loans.Loan_Status
```

```python
#knn = KNeighborsClassifier('n_neighbors=1')
model = LogisticRegression()
model.fit(X, y)
print model.predict_proba
```

## Making predictions using the model

```python
X_test = test_data[feature_cols]
predictions = model.predict(X_test)
predictions_proba = model.predict_proba(X_test)
print predictions
print predictions_proba
```

## Writing the submission file

```python
loan_id = test_data.Loan_ID
predictionsS = pd.Series(predictions)
loan_status_predictions = predictionsS.apply(lambda x: 'Y' if x == 1 else 'N')
```

```python
with open('submission.csv', 'wb') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(['Loan_ID', 'Loan_Status'])
    for i in xrange(len(loan_id)):
        spamwriter.writerow([loan_id[i], predictionsS[i]])
```
