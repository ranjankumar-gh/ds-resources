```python
import pandas as pd
import graphlab as gl
```

```python
#loans = pd.read_csv('train_u6lujuX.csv')
loans = gl.SFrame('train_u6lujuX.csv')
```

```python
loans
```

```python
features = ['Gender',              
            'Married',               
            'Dependents',     
            'Education', 
            'Self_Employed', 
            'ApplicantIncome',
            'CoapplicantIncome',
            'LoanAmount',
            'Loan_Amount_Term',
            'Credit_History',
            'Property_Area'
           ]
loans['safe_loans'] = loans['Loan_Status'].apply(lambda x : +1 if x=='Y' else -1)
loans.remove_column('Loan_Status')
target = 'safe_loans'
loans = loans[features + [target]]
```

```python
loans
```

```python
print "Total loan applications: ", len(loans)
print "No. of missing gender info: ", len(loans) - (len(loans[loans['Gender'] == 'Male']) + len(loans[loans['Gender'] == 'Female']))
len(loans[loans['Gender'] == ''])
```

### Balancing data

```python
safe_loans_raw = loans[loans[target] == 1]
risky_loans_raw = loans[loans[target] == -1]

# Undersample the safe loans.
percentage = len(risky_loans_raw)/float(len(safe_loans_raw))
risky_loans = risky_loans_raw
safe_loans = safe_loans_raw.sample(percentage, seed=1)
loans_data = risky_loans_raw.append(safe_loans)

print "Percentage of safe loans                 :", len(safe_loans) / float(len(loans_data))
print "Percentage of risky loans                :", len(risky_loans) / float(len(loans_data))
print "Total number of loans in our new dataset :", len(loans_data)
```

```python
for feature in features:
    loans_data_one_hot_encoded = loans_data[feature].apply(lambda x: {x: 1})    
    loans_data_unpacked = loans_data_one_hot_encoded.unpack(column_name_prefix=feature)
    
    # Change None's to 0's
    for column in loans_data_unpacked.column_names():
        loans_data_unpacked[column] = loans_data_unpacked[column].fillna(0)

    loans_data.remove_column(feature)
    loans_data.add_columns(loans_data_unpacked)
```
